## Identifiers, Variables, Data Types

### Identifiers
Any name we give to identify a **variable**, **function**, **class**, **module** or other **object**

**Rules of Identifiers**
1. A variable name must start with a letter or the underscore character
2. A variable name cannot start with a number
3. A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ )
4. Variable names are case-sensitive (age, Age and AGE are three different variables)
5. A variable name cannot be any of the Python keywords.

### Variables

Variables are containers for storing data values.

- Python has no command for declaring a variable.
- A variable is created the moment you first assign a value to it.

```python
# x = 12
# y = "Amide"
# print(x)
# print(y)
x = 12
print(type(x))
x = "Amide"
print(type(x))

# Python is a dynamic typed programming langauge
```

Variables do not need to be declared with any particular type, and can even change type after they have been set.

```python
x = 7     
x = "Rakesh"
print(x)
```

Python allows you to assign values to multiple variables in one line

```python
x, y, z = "Orange", "Banana", "Cherry"
print(x)
print(y)
print(z)
```

```python
x = 12
y = 19
# swap the number 
x, y =  y, x
```

And you can assign the same value to multiple variables in one line

```python
x = y = z = "Orange"
print(x)
print(y)
print(z)
```

Unpack a Collection

```python
#unpack of list
fruits = ["apple", "banana", "cherry"]
x, y, z = fruits
print(x)
print(y)
print(z)
```

**Global Varible**

## Strings

Strings in python are surrounded by either single quotation marks, or double quotation marks.

**'hello world'** is the same as **"hello world"**

```python
# String varibale
name = "Amide"
# name = 'Amide'
print(name)
```

You can assign a multiline string to a variable by using three quotes:

```python
# x  = "Hello world Hello world Hello world Hello world Hello world
# Hello world Hello world Hello world"
```

```python
# Mulitine String
a = """Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididunt
ut labore et dolore magna aliqua."""
print(a)
```

```python
# single quote support multiline
a = '''Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididunt
ut labore et dolore magna aliqua.'''
print(a)
```

String as Array

```python
# Get the character at position
message = "Hello world"
print(message[1])
print(message[2])
```

```python
#Looping Through a String
for n in "Apple":
  print(n)
```

```python
# The len() function returns the length of a string
message = "Hello world"
len(message)
```

**Slicing**

```python
# Get the characters from position 2 to position 5 (not included):
name = "Hello, World!"
print(name[3:5:2])
```

```python
# Slice From the Start
name = "Hello, World!"
print(name[:5])
```

**Modify Strings**

```python
#The upper() method returns the string in upper case:
a = "Hello, World!"
print(a.upper())
```

```python
# The lower() method returns the string in lower case:
a = "Hello, World!"
print(a.lower())
```

```python
# Remove Whitespace
a = " Hello, World    "
print(len(a.strip()))
```

```python
# Concatenation
x = "Hello"
# y = "World"
y = str(1) # str(1)
print(x + y)
#other methods, join, %, {}
```

```python
# format() method to insert numbers into strings:
job = "Developer"
txt = "My name is Amin, and I am {}"
print(txt.format(job))
```

```python
# The format() method takes unlimited number of arguments,
quantity = 3
itemno = 567
price = 49.95
myorder = "I want {} pieces of item {} for {} dollars."
print(myorder.format(quantity, itemno, price))
```

```python
# You can use index numbers : {1}
quantity = 3
itemno = 567
price = 49.95
myorder = "I want to pay {2} dollars for {0} pieces of item {1}."
print(myorder.format(quantity, itemno, price))
```

##  Boolean Expression and Conditional Statement

Booleans represent one of two values: True or False.

```python
#examples
print(10 > 9)
print(10 == 9)
print(10 < 9)
```

we can use boolean in our conditional statement

### if stament

```python
# Voter example
age = 16
if age >= 18:
   print("You are a voter")
```

### else statment

```python
# Voter example  
age = 17
if age >=18:
  print("Your are voter")
else:
  print("You are not Voter")
```

### elif: else if

```python
# education example
classNo = 8
if classNo <= 8:
  print("School")
elif classNo > 12:
  print("College")
else:
  print("High School")
```
