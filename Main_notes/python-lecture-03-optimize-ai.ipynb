## Dictionary

Dictionaries are used to store data values in key:value pairs.

```python
#examples of dict
data = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print(data)
```

```python
#Dictionary items are presented in key:value pairs, and can be referred to by using the key name
data = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print(data['brand'])
```

```python
#Dictionaries cannot have two items with the same key:
data = {
  "brand": "Ford",
  "model": "Mustang",
  "model": "swift",
  "year": 1964,
  "year": 2020
}
print(data)
```

```python
# number of items in the dictionary:
len(data)
```

```python
# The values in dictionary items can be of any data type:
data = {
  "brand": "Ford",
  "electric": False,
  "year": 1964,
  "colors": ["red", "white", "blue"]
}
print(data['colors'][1])
```

```python
#type()
print(type(data))
```

```python
# Using the dict() method to make a dictionary:
data = dict(name = "John", age = 36, country = "Norway")
print(data)
```

#### Accessing Items

```python
# Get the value of the "model" key:
data = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
x = data["model"]
print(x)
```

```python
# There is also a method called get()
x = data.get("model")
print(x)
```

```python
#get is best over ['']
print(data["models"])
print(data.get('models'))
```

```python
# Get a list of the keys:
x = data.keys()
print(x)
```

```python
#keys function return always updated keys
car = {
"brand": "Ford",
"model": "Mustang",
"year": 1964
}

x = car.keys()

print(x) #before the change #1

car["color"] = "white"

print(x) #after the change #2
```

```python
# Get a list of the values:
x = data.values()
print(x)
```

The list of the values is a view of the dictionary, meaning that any changes done to the dictionary will be reflected in the values list same as keys

```python
car = {
"brand": "Ford",
"model": "Mustang",
"year": 1964
}

x = car.values()

print(x) #before the change

car["year"] = 2020

print(x) #after the change
```

```python
# Get a list of the key:value pairs
x = data.items()
print(x)
```

```python
#check the list
data = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
if "model" in data:
  print("Yes, 'model' is one of the keys in the data dictionary")
```

```python
#change value 
data = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
data["year"] = 2018
```

```python
#change value using update function
data.update({'year': 2000})
print(data)
```

```python
#add items
data = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
data["color"] = 'red'
```

```python
#add items using update
data = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
data.update({"color": 'blue'})
```

```python
# removes the item
data = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
data.pop("model")
print(data)
```

```python
# popitem() method removes the last inserted item
data = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
data.popitem()
print(data)
```

```python
# The del keyword removes the item with the specified key name
data = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
del data["model"]
print(data)
```

```python
# The clear() method empties the dictionary
data = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
data.clear()
print(data)
```

```python
data = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
```

```python
# Print all key names in the dictionary
for x in data:
  print(x)
```

```python
# Print all values in the dictionary
for x in data:
  print(data[x])
```

```python
# Print all values in the dictionary
for x in data.values():
  print(x)
```

```python
# You can use the keys() method to return the keys of a dictionary
for x in data.keys():
  print(x)
```

```python
# loop through both keys and values
for x, y in data.items():
  print(x, y)
```

 __init__() Function 

```python
# Note: The __init__() function is called automatically every time the class is being used to create a new object.
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

p1 = Person("John", 36)

print(p1.name)
print(p1.age)
```

```python
#We can also define a function inside a Python class
# create a class
class Room:
    length = 0.0
    breadth = 0.0
    
    # method to calculate area
    def calculate_area(self,):
        print("Area of Room =", self.length * self.breadth)

# create object of Room class
study_room = Room()

# assign values to all the attributes 
study_room.length = 42.5
study_room.breadth = 30.8

# access method inside class
study_room.calculate_area()
```

```python
#We can also define a function inside a Python class
# create a class
class Room:

    def __init__(self, l, b):
      self.length = l
      self.breadth = b
    
    # method to calculate area
    def calculate_area(self):
        print("Area of Room =", self.length * self.breadth)

# create object of Room class
study_room = Room(20, 10)

# # assign values to all the attributes 
# study_room.length = 42.5
# study_room.breadth = 30.8

# access method inside class
study_room.calculate_area()
```

```python
class Node:
  def __init__(self, v):
    self.value = v
    self.next = None
```

```python
n1 = Node(3)
n2 = Node(4)
n3 = Node(5)
```

```python
print(n1.value)
```

```python
print(n2.value)
```

```python
print(n1.next)
```

```python
n1.next = n2
n2.next = n3
n3.next = n1
```

```python
print(n1.next.value)
```

```python
print(n1.next.next.value)
```

```python
print(n1.next.next.next.next.value)
```
